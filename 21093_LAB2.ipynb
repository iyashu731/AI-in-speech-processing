```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import IPython.display as ipd

y, sr = librosa.load('speechlearning.wav')
first_derivative = np.diff(y)
first_derivative /= np.max(np.abs(first_derivative))

print("Playing an audio First Derivative Signal:")

plt.figure(figsize=(10, 5))
librosa.display.waveshow(first_derivative, sr=sr, color='green')
plt.title('FIRST DERIVATIVE')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()

ipd.Audio(first_derivative, rate=sr)
```

```python
plt.figure(figsize=(10, 5))
librosa.display.waveshow(y, sr=sr,color='red')
plt.title('Original')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()

ipd.Audio(y, rate=sr)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data for demonstration
derivative_1 = np.random.randn(1000)  

# Calculation of zero crossings and regions
zero_crossing = np.where(np.diff(np.sign(derivative_1)))[0]
diff = np.diff(zero_crossing)
threshold = 1000
speech_regions = diff[diff > threshold]
silence_regions = diff[diff <= threshold]

# Calculating average lengths
avg_length_speech = np.mean(speech_regions)
avg_length_silence = np.mean(silence_regions)

# Printing average lengths
print(f"Average length between consecutive zero crossings in speech regions: {avg_length_speech}")
print(f"Average length between consecutive zero crossings in silence regions: {avg_length_silence}")

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(diff, label='All regions', color='blue')
plt.plot(np.arange(len(speech_regions)), speech_regions, 'ro', label='Speech regions', color='red')
plt.plot(np.arange(len(speech_regions), len(speech_regions) + len(silence_regions)), silence_regions, 'bo', label='Silence regions', color='green')
plt.title('Pattern of Zero Crossings')
plt.xlabel('Zero Crossing Difference')
plt.ylabel('Difference between Consecutive Zero Crossings')
plt.legend()
plt.show()

# Printing patterns
print("Pattern of the Zero Crossings:\n"
      f"All regions: {diff}\n"
      f"Speech regions: {speech_regions}\n"
      f"Silence regions: {silence_regions}")
```

```python
my_word_files = ['cricket 1.mpeg', 'virat1.mpeg', 'msdhoni1.mpeg', 'kadapa 1.mpeg', 'sl1.mpeg']
teammate_word_files = ['cricket.wav', 'virat.wav', 'msdhoni.wav', 'kadapa.wav', 'sl.wav']
words = ['cricket', 'virat', 'msdhoni', 'kadapa', 'sl']
my_word_lengths = []
teammate_word_lengths = []

for word_file in my_word_files:
    signal, sr = librosa.load(word_file, sr=None)
    length_seconds = len(signal) / sr
    my_word_lengths.append(length_seconds)

for word_file in teammate_word_files:
    signal, sr = librosa.load(word_file, sr=None)
    length_seconds = len(signal) / sr
    teammate_word_lengths.append(length_seconds)

print("Lengths of the spoken words MINE:", my_word_lengths)
print("Lengths of the spoken words TeamMate:", teammate_word_lengths)

bar_width = 0.35
index = np.arange(len(words))
plt.figure(figsize=(12, 6))
plt.bar(index - bar_width/2, my_word_lengths, bar_width, label='My Words', color='green')
plt.bar(index + bar_width/2, teammate_word_lengths, bar_width, label="Teammate's Words", color='purple')
plt.xlabel('Words')
plt.ylabel('Length (seconds)')
plt.title('Comparison between Spoken Words Length')
plt.xticks(index, words)
plt.legend()

plt.show()
```
